```python
import os
import ast

def tick_to_time_us(tick):
    return (tick * (1/50e6)) * 1e6

    
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
num_samples = 20

# Generate sample numbers from 1 to 20
samples = np.arange(1, num_samples + 1)

# Generate switch delay values for each operating system
free_rtos_v = np.full(num_samples, 17)  # FreeRTOS
nutex_v = np.full(num_samples, 21)  # NuttX
zephyr_v = np.full(num_samples, 24)  # Zephyr

# Create a line plot with thicker lines and mark each sample with a point
plt.figure(figsize=(10, 6))
plt.plot(samples, free_rtos_v, 'o--', label='FreeRTOS', linewidth=2)
plt.plot(samples, nutex_v, 'o--', label='NuttX', linewidth=2)
plt.plot(samples, zephyr_v, 'o--', label='Zephyr', linewidth=2)

plt.xlabel('Numer próbki')
plt.ylabel('Czas (us)')
plt.title('Czas przełączenia kontekstu dla różnych systemów operacyjnych')

# Set the limits of the Y-axis and X-axis
plt.ylim(0, 30)
plt.xlim(0, 25)

# Set the x-ticks
plt.xticks(np.arange(1, num_samples + 1, 2))

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the operating systems
os = ['FreeRTOS', 'NuttX', 'Zephyr']

# Define the task creation times for each operating system
creation_times = [10, 20, 30]

# Define the task deletion times for each operating system
deletion_times = [15, 25, 35]

# Define the bar width
bar_width = 0.35

# Define the positions of the bars
r1 = np.arange(len(os))
r2 = [x + bar_width for x in r1]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(r1, creation_times, width=bar_width, align='center', label='Czas tworzenia')
plt.bar(r2, deletion_times, width=bar_width,  align='center', label='Czas usuwania')

# Add labels and title
plt.xlabel('System operacyjny')
plt.ylabel('Czas (us)')

# Add xticks on the middle of the group bars
plt.xticks([r + bar_width / 2 for r in range(len(os))], os)

# Create legend
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the operating systems
os = ['FreeRTOS', 'NuttX', 'Zephyr']

# Define the task creation times for each operating system
creation_times_run_block = [tick_to_time_us(3123), 20, 30]
creation_times_redy_to_start = [tick_to_time_us(3079), 20, 30]
creation_times_never_run = [tick_to_time_us(3222), 20, 30]

# Define the task deletion times for each operating system
deletion_times_run_block = [tick_to_time_us(1500), 25, 35]
deletion_times_redy_to_start = [tick_to_time_us(1500), 25, 35]
deletion_times_never_run = [tick_to_time_us(1500), 25, 35]

# Define the bar width
bar_width = 0.35

# Define the positions of the bars
r1 = np.arange(len(os))
r2 = [x + bar_width for x in r1]

# Create the figure
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Create the first subplot
axs[0].bar(r1, creation_times_run_block, width=bar_width, align='center', label='Czas tworzenia')
axs[0].bar(r2, deletion_times_run_block, width=bar_width,  align='center', label='Czas usuwania')
axs[0].set_xlabel('System operacyjny')
axs[0].set_ylabel('Czas (us)')
axs[0].set_xticks([r + bar_width / 2 for r in range(len(os))])
axs[0].set_xticklabels(os)
axs[0].legend()

# Create the second subplot
axs[1].bar(r1, creation_times_redy_to_start, width=bar_width, align='center', label='Czas tworzenia')
axs[1].bar(r2, deletion_times_redy_to_start, width=bar_width,  align='center', label='Czas usuwania')
axs[1].set_xlabel('System operacyjny')
axs[1].set_ylabel('Czas (us)')
axs[1].set_xticks([r + bar_width / 2 for r in range(len(os))])
axs[1].set_xticklabels(os)
axs[1].legend()

# Create the third subplot
axs[2].bar(r1, creation_times_never_run, width=bar_width, align='center', label='Czas tworzenia')
axs[2].bar(r2, deletion_times_never_run, width=bar_width,  align='center', label='Czas usuwania')
axs[2].set_xlabel('System operacyjny')
axs[2].set_ylabel('Czas (us)')
axs[2].set_xticks([r + bar_width / 2 for r in range(len(os))])
axs[2].set_xticklabels(os)
axs[2].legend()

# Show plot
plt.tight_layout()
plt.show()
```
